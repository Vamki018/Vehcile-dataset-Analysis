### 3. Fix Inconsistencies in Categorical Data

* Categorical inconsistencies fixed by:

    * Standardizing text case

    * Removing extra spaces

    * Merging equivalent categories (e.g., Petrol vs Gasoline)

* Resulted in clean and uniform categories

```python
cat_cols = [
    'fuel', 'transmission', 'body', 'drivetrain',
    'make', 'model', 'trim',
    'exterior_color', 'interior_color'
]


for col in cat_cols:
    df[col] = df[col].astype(str).str.strip().str.title()
```

```python
df['fuel'] = df['fuel'].replace({
    'Petrol': 'Petrol',
    'Gasoline': 'Petrol',
    'Diesel': 'Diesel',
    'Electric': 'Electric',
    'Hybrid': 'Hybrid'
})


df['fuel'].value_counts()
```

### 4. Convert Data Types

* Numerical columns converted to appropriate integer types

* Categorical columns converted to category data type

* Corrections ensured:

    * Accurate calculations

    * Efficient grouping operations

    * Reduced memory usage


```python
#Converting the data types of numerical columns
```

```python
df['year'] = df['year'].astype(int)
df['price'] = df['price'].astype(int)
df['mileage'] = df['mileage'].astype(int)
df['cylinders'] = df['cylinders'].astype(int)
df['doors'] = df['doors'].astype(int)
```

```python
#Converting the datatype of categorical colums
```

```python
cat_cols = [
    'fuel', 'transmission', 'body', 'drivetrain',
    'make', 'model', 'trim','type',
    'exterior_color', 'interior_color'
]

for col in cat_cols:
    df[col] = df[col].astype('category')
```

### Bivariate Analysis (Two Columns)

```python
sns.boxplot(x='doors', y='price', data=df)
plt.title("Price Distribution by Number of Doors")
plt.show()
```

* 2-door vehicles show higher median prices with wider variance.

* 4-door vehicles cluster around mid-range pricing, indicating mass-market positioning.

```python
#price vs mileage
sns.regplot(x='mileage', y='price', data=df)
plt.title("Mileage vs Price")
plt.show()
```

* Clear negative relationship

* Higher mileage → lower price

```python
#fuel vs price
sns.boxplot(data = df,y='price',x='fuel')
plt.xlabel('fuel')
plt.xticks(rotation = 45)
plt.ylabel('price')
plt.title('Price Distribution by Fuel Type')
plt.show()
```

* Electric and diesel vehicles have higher median prices than petrol variants.

* Fuel type significantly influences pricing strategy.

* Variance within categories suggests brand-level effects.

```python
#body vs price
sns.boxplot(data=df, x='body', y='price')
plt.xticks(rotation=45)
plt.title('Price Distribution by Body Type')
plt.show()
```

* SUVs and trucks show higher median prices, reflecting their utility, size, and feature-rich positioning.

* Sedans and hatchbacks cluster in the mid to lower price range, indicating mass-market targeting.


```python
pd.crosstab(df['make'], df['fuel']).plot(kind='bar', stacked=True)
plt.title("Brand vs Fuel Type")
plt.show()
```

* Certain brands specialize in specific fuel types.

* Market positioning differs clearly across brands.

* Indicates segmented product strategies.

```python
#body vs drivetrain
sns.countplot(data=df, x='body', hue='drivetrain')
plt.xticks(rotation=45)
plt.title('Body Type vs Drivetrain')
plt.show()
```

* SUVs and trucks are dominated by AWD/4WD drivetrains, reflecting their design for off-road capability and load handling.

* Sedans and hatchbacks are largely FWD, optimized for efficiency and urban use.

* RWD appears in limited body types, typically sports cars or premium sedans.

### Multivariate Analysis

```python
#Correlation Heatmap
num_cols = ['price', 'mileage', 'year', 'cylinders', 'doors']

sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

* Price shows a strong positive correlation with year and a negative correlation with mileage, confirming that newer, less-used vehicles retain higher value.

* cylinders has a moderate positive relationship with price, while domors shows weak correlation, indicating that engine performance influences pricing more than body configuration.

* The strong negative correlation between year and mileage highlights potential multicollinearity, which should be handled carefully in predictive models.
